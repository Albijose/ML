```python
import pandas as pd
from  matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data =pd.read_csv("homeprices.csv")
data
```

```python
plt.xlabel('area in sqft')
plt.ylabel('price in us$')
plt.scatter(data.area,data.price,marker='*',color='Blue')
```

```python
x=data.drop('price',axis=1)#features
y=data['price']#target
```

```python
x
```

```python
y
```

```python
our_model=LinearRegression()#creating object of the linaer regression
our_model.fit(x,y)#fit=model training
```

```python
our_model.coef_
```

```python
y_pred=our_model.predict(x)#by using x value to predict y
```

```python
pd.DataFrame({'true_y_values':y,'predicted_y_values':y_pred})
```

```python
plt.scatter(x,y)
plt.xlabel('area')
plt.ylabel('predicted price')
plt.plot(data['area'],y_pred,color='red')
```

```python
our_model.predict([[3700]])
```

```python
data=pd.read_csv('canada_per_capita_income.csv')
data
```

```python
plt.xlabel('Year')
plt.ylabel('PreCapitaIncome')
plt.scatter(data.Year,data.PerCapitaIncome,marker='p',color='Yellow')
```

```python
x=data.drop('PerCapitaIncome',axis=1)
y=data.PerCapitaIncome
```

```python
x
```

```python
y
```

```python
our_model=LinearRegression()#creating object of the linaer regression
our_model.fit(x,y)#fit=model training
```

```python
our_model.coef_
```

```python
our_model.intercept_
```

```python
y_pred=our_model.predict(x)
```

```python
our_model.predict([[2020]])
```
